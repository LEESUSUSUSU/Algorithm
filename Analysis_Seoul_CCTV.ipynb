```python
from os import rename
CCTV_Seoul= CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] : '구별'})
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel("/content/drive/MyDrive/zerobase/ds_study/data/Seoul_Population.xlsx", header=2, usecols="B,D,G,J,H")
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={
    pop_Seoul.columns[0] : '구별',
    pop_Seoul.columns[1] : '인구수',
    pop_Seoul.columns[2] : '한국인',
    pop_Seoul.columns[3] : '외국인',
    pop_Seoul.columns[4] : '고령자'
}, inplace=True)
pop_Seoul.head()
```

```python
CCTV_Seoul["최근증가율"] = ((CCTV_Seoul["2016년"]+ CCTV_Seoul["2015년"]+CCTV_Seoul["2014년"])/ CCTV_Seoul["2013년도 이전"]*100)
CCTV_Seoul.sort_values(by="최근증가율", ascending=False).head(5)
```

단순히 구별 cctv 개수로는 한계가 있습니다. 인구현황과 함께 비교 해야할 것 같습니다.

```python
pop_Seoul.head()
```

```python
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul["구별"].unique()
```

```python
len(pop_Seoul["구별"].unique())
```

```python
pop_Seoul["외국인비율"] = pop_Seoul["외국인"] / pop_Seoul["인구수"] * 100
pop_Seoul["고령자비율"] = pop_Seoul["고령자"] / pop_Seoul["인구수"] * 100
pop_Seoul.head(5)
```

```python
pop_Seoul.sort_values(by="외국인", ascending=False).head(5)
```

```python
pop_Seoul.sort_values(by="고령자", ascending=False).head(5)
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on="구별")
data_result.head()
```

```python
del data_result["2013년도 이전"]
del data_result["2014년"]
del data_result["2015년"]
del data_result["2016년"]
data_result.head()
```

```python
data_result.set_index("구별", inplace=True)
data_result.head()
```

```python
data_result.corr()
```

```python
data_result["cctv비율"] = data_result["소계"] / data_result["인구수"]
data_result["cctv비율"] = data_result["cctv비율"] *100
data_result.sort_values(by="cctv비율", ascending=False).head(5)
# %
```
